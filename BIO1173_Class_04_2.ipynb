## **Multimodal Embedding**

------------------

In this lesson, we explore what embeddings are why they play a special role in complex, highly integrated academic activites such as UT San Antonio's **One Health Program**.


#### **What is an Embedding?**
At its core, an **embedding** is a way of representing data—whether it’s a word, an image, or a sound—as a point in a high-dimensional mathematical space. These points are represented as long lists of numbers called **vectors**.

The "magic" of embeddings is that they capture **semantic meaning**. In an embedding space, items that are conceptually similar are placed close together, while unrelated items are placed far apart. For example, in a text embedding, the words "piglet" and "hog" would be mathematically closer to each other than the words "piglet" and "transmission."

#### **Why It Matters for the "One Health" Paradigm**
For students and faculty participating the One Health Program, this technology allows them to analyze complex, unstructured data at a scale never before possible. By converting "real world" observations into high-dimensional vectors, it becomes possible to build systems that "understand" the compled relationships between Human Health, Animal Health and Environmental Health.

#### **Example Use Cases**

Below are three use cases where these multimodal embeddings could provide a unique academic and research advantage.

##### **1. Rapid Zoonotic Disease Mapping (The "Image-to-Insight" Bridge)**
In One Health, identifying how a disease moves from wildlife to humans is critical. Students can use multimodal embeddings to link visual field data with genomic sequences.

* **The Advantage:** Instead of searching by keywords like "bat habitat," a student can upload a **drone photo** of a specific limestone cave in the Texas Hill Country.
* **The Workflow:** The embedding model maps the visual features of that environment to the same vector space as **clinical reports** and **pathogen genomic data** stored in BIO 1173 databases.
* **Outcome:** Faculty can instantly find research papers or genomic sequences that match the specific environmental conditions shown in a photograph, bypassing the need for manual tagging.

##### **2. Longitudinal Environmental Health Monitoring via Video**
The BHE department focuses heavily on environmental impacts like the health of the Edwards Aquifer or urban expansion in San Antonio.

* **The Advantage:** Researchers can embed **time-lapse videos** of a local watershed alongside **chemical sensor data** (pH levels, nitrogen content).
* **The Workflow:** By embedding video frames, the model can "cluster" moments where water turbidity (visual) matches spikes in bacterial contamination (text/data).
* **Outcome:** Students in the **EcoJEDI** or **Tuggle Scholars** programs can query their data by asking: *"Find me moments in the video stream that look like the contamination event from last July,"* and the model will retrieve the relevant video segments based on visual similarity.

##### **3. Cross-Disciplinary "BHE CUREs" Knowledge Retrieval**
With different labs working on everything from **Bacteriophages** to **Medicinal Honey**, information can become siloed.

* **The Advantage:** Multimodal embeddings allow for a "Universal Search" across the entire department’s output.
* **The Workflow:** A student working on the **HONEY Pathway** can upload a **micrograph image** of a specific bacterial inhibition zone.
* **Outcome:** The embedding model can find a match in the **Bacteriophage Discovery** lab's text-based research posters because the "visual signature" of the bacterial stress in the image is mathematically similar to the "textual description" of stress in the phage lab’s database.

## Create Helper Functions

Here are reusable functions that show exactly how to encode each content type and get embeddings.

````python
# @title Create Helper Functions

def embed_text(text: str) -> np.ndarray:
    """Encode a text string into an embedding vector.

    Simply pass the string directly to embed_content.
    """
    result = client.models.embed_content(
        model=MODEL_ID,
        contents=[text],
    )
    return np.array(result.embeddings[0].values)


def embed_image(image_path: str) -> np.ndarray:
    """Encode an image file (PNG or JPEG) into an embedding vector.

    Read the file as bytes and wrap it in types.Part.from_bytes
    with the appropriate MIME type.
    """
    # Determine MIME type from extension
    ext = image_path.lower().rsplit('.', 1)[-1]
    mime_map = {'png': 'image/png', 'jpg': 'image/jpeg', 'jpeg': 'image/jpeg'}
    mime_type = mime_map.get(ext, 'image/png')

    with open(image_path, 'rb') as f:
        image_bytes = f.read()

    result = client.models.embed_content(
        model=MODEL_ID,
        contents=[
            types.Part.from_bytes(data=image_bytes, mime_type=mime_type),
        ],
    )
    return np.array(result.embeddings[0].values)


def embed_audio(audio_path: str) -> np.ndarray:
    """Encode an audio file (WAV or MP3) into an embedding vector.

    Read the file as bytes and wrap it in types.Part.from_bytes
    with the appropriate MIME type. Max duration: 80 seconds.
    """
    ext = audio_path.lower().rsplit('.', 1)[-1]
    mime_map = {'wav': 'audio/wav', 'mp3': 'audio/mp3'}
    mime_type = mime_map.get(ext, 'audio/wav')

    with open(audio_path, 'rb') as f:
        audio_bytes = f.read()

    result = client.models.embed_content(
        model=MODEL_ID,
        contents=[
            types.Part.from_bytes(data=audio_bytes, mime_type=mime_type),
        ],
    )
    return np.array(result.embeddings[0].values)


print("Encoding functions ready!")
print(f"  embed_text('hello')   -> {MODEL_ID} text embedding")
print(f"  embed_image('x.png')  -> {MODEL_ID} image embedding")
print(f"  embed_audio('x.wav')  -> {MODEL_ID} audio embedding")
````

## **Significance of Multimodal Embedding Alignment**

This code demonstrates the most powerful feature of the **Gemini-Embedding-2-Preview** model: **Joint Vector Space Alignment**. In a One Health context, this is what allows a computer to "understand" that a written description of a lemur, a photo of a lemur, and the sound of a lemur all represent the same biological entity.

#### **1. Dimensionality Uniformity (The 3072-D Vector)**
As seen in your output, every modality—regardless of its original format (PNG, JPG, WAV)—is compressed into a vector of exactly **3072 numbers**.
* **Why it matters:** In traditional computing, you cannot compare a pixel to a sound wave. By converting them both into 3072-dimensional vectors, you turn "sights" and "sounds" into "coordinates." You can now use standard geometry (like cosine similarity) to calculate how "close" an image is to an audio file.

#### **2. Semantic Mapping (Same Space Alignment)**
The phrase *"in the same space"* is the key takeaway. It means the model has been trained such that the vector for the word "Roar" and the vector for the actual audio file of a roar are mathematically near each other.

* **Significance for BIO 1173:** Students don't have to write complex code to translate audio to text. They can simply search their database using a "Common Currency."
    * **Example:** A student can query the system with a **text** string ("Aggressive primate vocalization") and the system can retrieve the **audio** file `RoarShriek.wav` because their vectors share the same neighborhood in that 3072-dimensional space.

#### **3. One Health Application: Cross-Modal Research**
This code is the "engine" behind the use cases we discussed earlier. It enables the following capabilities:

| Activity | Input Modality | Target Modality | One Health Use Case |
| :--- | :--- | :--- | :--- |
| **Search** | **Text** ("Invasive species") | **Image** | Finding drone photos of specific invasive plants. |
| **Analysis** | **Audio** (Forest soundscape) | **Text** | Automatically labeling species presence in a conservation area. |
| **Clustering** | **Image** (Lab slide) | **Data/Genomics** | Linking visual cellular damage to specific genetic mutations. |

This code proves that the **Department of Biology, Health, and the Environment (BHE)** can now treat disparate environmental data as a single, searchable library. It removes the technical barrier of "data types," allowing students to focus on the **biological relationships** between the samples.

## Example 5: Compute and Visualize Text-to-Image Similarity

The code in the cell below shows how to compute and visualize the similarity between embedded text and and embedded images.It represents the "Discovery Phase" of a **Computational Biology (BIO 1173)** project. Now that the data has been converted into numbers (i.e. vectorized), this script evaluates how well the AI "understands" the relationship between your labels and your media.

#### **1. The Mathematical Engine: Cosine Similarity**
The core of this cell is `cosine_similarity(text_embeddings, image_embeddings)`.
* **What it's doing:** It calculates the cosine of the angle between every text vector and every image vector.
* **The Scale:** A score of **1.000** means the vectors are identical (perfect conceptual match), while **0.000** means they are completely unrelated.

#### **2. The Visualization: The Heatmap**
The `sns.heatmap` function creates a $7 \times 7$ grid that visualizes the entire "One Health" dataset at once.
* **The "Hot" Diagonal:** If your labels and images are aligned (e.g., the 2nd text describes the 2nd image), you should see a bright red/orange diagonal line cutting through the heatmap.
* **Off-Diagonal Values:** Lower scores (yellow) represent the "noise"—how much the AI thinks an "Insulin signaling pathway" looks like a "Gray fox" (ideally, not much).

#### **3. The Inference: Automated Matching**
The final loop uses `np.argmax` to perform an automated search.
* **The Logic:** For every image (column), it looks for the row (text) with the highest similarity score.
* **One Health Significance:** This demonstrates how a researcher could use AI to automatically label thousands of field photos or lab results. Instead of a human manually tagging every photo of a **Bacteriophage**, the AI suggests the best label based on the 3072-dimensional "knowledge" it gained during embedding.

---

#### **Why this important?**

| Feature | Student Benefit |
| :--- | :--- |
| **Concept Validation** | Proves the AI "sees" what you see (e.g., matching "Water sampling" to the correct vial photo). |
| **Error Detection** | If the "Best Match" for a Lemur is "Bacteriophage," it alerts the student to a data error or model bias. |
| **Efficiency** | Moves from looking at individual files to analyzing a $7 \times 7$ matrix of relationships instantly.

The output shows the **Image Filename** $\rightarrow$ **Best Text Match** along with a confidence score. In the **One Health** paradigm, these scores provide a quantitative way to measure the "semantic distance" between different types of biological evidence.

````python
# @title Example 5: Compute and Visualize Text-to-Image Similarity

# Define colormap
COLORMAP = "YlOrRd"

# Cosine similarity: (num_texts x num_images)
text_image_sim_eg = cosine_similarity(text_embeddings_eg, image_embeddings_eg)

# Shorten labels for display
short_texts_eg = [t[:50] + "..." if len(t) > 50 else t for t in text_descriptions_eg]

plt.figure(figsize=(10, 6))
ax = sns.heatmap(
    text_image_sim_eg,
    annot=True,
    fmt=".3f",
    cmap=COLORMAP,
    xticklabels=image_labels_eg,
    yticklabels=short_texts_eg,
    vmin=0, vmax=1,
)
ax.set_xlabel("Images", fontsize=12)
ax.set_ylabel("Text Descriptions", fontsize=12)
ax.set_title("Text \u2192 Image Cosine Similarity", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Print best matches
print("\n\U0001f3af Best text match for each image:")
for j, img in enumerate(image_labels_eg):
    best_idx = np.argmax(text_image_sim_eg[:, j])
    print(f"  {img:15s} \u2192 \"{text_descriptions_eg[best_idx]}\" (score: {text_image_sim_eg[best_idx, j]:.3f})")
````

## Example 7: Image-to-Text Similarity (Reverse Search)

The code in the cell below performs a "Reverse Search" in which a text string is used to search for an image.

````python
# @title Example 7: Image-to-Text Similarity (Reverse Search)

# Search text
text_corpus_eg = [
    "Bacteriophage experiment",
    "Gray fox picture",
    "Insulin signaling pathway",
    "Black and White Ruffed Lemur",
    "Tiny Earth research project",
    "Roar shriek vocalization",
    "Environmental Water sampling",
    "Research Methods in Zoo Biology",
    "Bioacoustic Approaches Conservation",
    "Wildlife Biogeography & Environmental Health"
]

print("Reverse search image-to-text similarities...")
corpus_embeddings_eg = np.array([embed_text(t) for t in text_corpus_eg])

# For each image, find the top-2 most similar texts
image_corpus_sim_eg = cosine_similarity(image_embeddings_eg, corpus_embeddings_eg)

for i, img in enumerate(image_labels_eg):
    print(f"\n{'='*60}")
    display(IPImage(filename=image_files_eg[i], width=250))
    print(f"\U0001f50d Top 2 text matches for {img}:")
    top_indices = np.argsort(image_corpus_sim_eg[i])[::-1][:2]
    for rank, idx in enumerate(top_indices, 1):
        print(f"  {rank}. \"{text_corpus_eg[idx]}\" (score: {image_corpus_sim_eg[i, idx]:.3f})")
````

## **Full Cross-Modal HeatMap**

The code in Example 9 shows how to create a Cross-Modal HeatMap for all the stored embeddings. It represents the "Grand Synthesis" of the data analysis workflow. It takes every piece of evidence—regardless of whether it is a written description, a visual photo, or an acoustic recording—and evaluates the entire system's connectivity in a single mathematical structure.

#### **Data Aggregation (The "Unified Stack")**
The code uses `np.vstack` to vertically stack the text, image, and audio embeddings into one master matrix called `all_embeddings`.
* **The Logic:** Because all three modalities were converted into the same **3072-dimensional space**, they can be combined into a single table.
* **The Shape:** If you have 7 texts, 7 images, and 1 audio clip, your combined matrix becomes a **$15 \times 3072$** block of data.

#### **Universal Similarity Computation**
The command `full_sim = cosine_similarity(all_embeddings)` computes a **square matrix**. In this case, it is a $15 \times 15$ grid where every single item is compared against every other item (including itself).
* **The "Self-Match":** The diagonal of this matrix will always be **1.00**, as every item is perfectly similar to itself.
* **The Cross-Modal Insights:** This matrix reveals "hidden" relationships, such as how closely a wildlife photo (Image) relates to a specific forest sound (Audio), even if they weren't explicitly labeled as a pair.

#### **Visualization: The Viridis Heatmap**
Using the `viridis` color map, the code creates a dense, informative visualization of the entire dataset's relationships.
* **Labeling:** It intelligently adds emojis (📝 for text, 🖼️ for images, 🎵 for audio) to the axes so the researcher can quickly see which modality is being compared.
* **Symmetry:** The heatmap is perfectly symmetrical across the diagonal.

## Example 9: Full Cross-Modal HeatMap

The code in the cell below generate a full, cross-model heatmap of all of the assets used in the Examples.

````python
# @title Example 9: Full Cross-Modal HeatMap

import warnings

# Define colormap
COLORMAP = "viridis"

# Suppress font-related UserWarnings
warnings.filterwarnings("ignore", category=UserWarning, message=".*Glyph.*missing from font.*")

# Combine all embeddings and labels
all_labels_eg = (
    [f"\U0001f4dd {t[:40]}..." if len(t) > 40 else f"\U0001f4dd {t}" for t in text_descriptions_eg]
    + [f"\U0001f5bc\ufe0f {img}" for img in image_labels_eg]
    + [f"\U0001f3b5 {aud}" for aud in audio_labels_eg]
)

all_embeddings_eg = np.vstack([
    text_embeddings_eg,
    image_embeddings_eg,
    audio_embeddings_eg,
])

print(f"Combined embedding matrix: {all_embeddings_eg.shape}")
print(f"Labels: {len(all_labels_eg)} items")

# Full cosine similarity matrix
full_sim_eg = cosine_similarity(all_embeddings_eg)

plt.figure(figsize=(14, 10))
ax = sns.heatmap(
    full_sim_eg,
    annot=True,
    fmt=".2f",
    cmap=COLORMAP,
    xticklabels=all_labels_eg,
    yticklabels=all_labels_eg,
    vmin=0, vmax=1,
    square=True,
)
ax.set_title("Full Cross-Modal Similarity (Text + Image + Audio)", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
````
